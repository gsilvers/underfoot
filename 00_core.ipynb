```
#export
def say_hello(to):
    "Say hello to somebody"
    return f'Hello {to}!'
```

```
#hide_output
#hide_input

import pygame
import time 
import sys
import os

displays_valid = 0
pygame.display.list_modes()
if not pygame.display.list_modes():
    displays_valid = 1

if displays_valid == 0:    
    pygame.init()
    screen = pygame.display.set_mode([500, 500])
    running = True
    i = 0
    count_direction = 0
    while running:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_q:
                    pygame.draw.circle(screen, (200, 200, 10), (200,200), 200)
                    pygame.display.flip()
                    time.sleep(5)
                    pygame.quit()
                    sys.exit()
        screen.fill((255, 255, 255))
        pygame.draw.circle(screen, (33, i, 50), (i, i), 75)
        pygame.display.flip()
        time.sleep(.1)
        if count_direction == 0:
            if i >= 248:
                count_direction = 1
            i = i + 2
        if count_direction == 1:
            if i == 2:
                count_direction = 0   
            i = i - 2
    pygame.quit()
```
