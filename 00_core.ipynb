# Underfoot

> Underfoot is a 2D game built on top of the pygame library about a dog and her human best friend. This game is developed on top of NBDev and Jupyter. To start I am simply running throught the NBDev Tutorial and maybe testing some pygame.

```
#export
def say_hello(to):
    "Say hello to somebody"
    return f'Hello {to}!'
```

```
#export
class HelloSayer:
    "Say hello to `to` using `say_hello`"
    def __init__(self, to): self.to = to

    def say(self):
        "Do the saying"
        return say_hello(self.to)
```

```
#hide_input
#hide_outputmake
import pygame
import time 
import sys

pygame.init()
screen = pygame.display.set_mode([500, 500])
running = True
i = 0
count_direction = 0
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_q:
                pygame.draw.circle(screen, (200, 200, 10), (200,200), 200)
                pygame.display.flip()
                time.sleep(5)
                pygame.quit()
                sys.exit()
    screen.fill((255, 255, 255))
    print("i: ", i)
    pygame.draw.circle(screen, (33, i, 50), (i, i), 75)
    pygame.display.flip()
    time.sleep(.1)
    if count_direction == 0:
        if i >= 248:
            count_direction = 1
        i = i + 2
    if count_direction == 1:
        if i == 2:
            count_direction = 0   
        i = i - 2
pygame.quit()
```
